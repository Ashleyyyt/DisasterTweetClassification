# Disaster Tweets Machine Learning Model

### EDA

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load data
df_train = pd.read_csv("Data/train.csv")
df_test = pd.read_csv("Data/test.csv")
print(df_train.head(10))
print(df_test.head(10))
```

```python
print('Training Set Shape = {}'.format(df_train.shape))
print('Test Set Shape = {}'.format(df_test.shape))
```

```python
print("Training Set Nulls:")
print(df_train.isna().sum())
print("\nTest Set Nulls:")
print(df_test.isna().sum())

fig, axes = plt.subplots(ncols=2, figsize=(17, 4), dpi=100)

# Manually set missing val cols
missing_cols = ['keyword', 'location']
sns.barplot(x=df_train[missing_cols].isnull().sum().index, y=df_train[missing_cols].isnull().sum().values, ax=axes[0], palette='Set2')
sns.barplot(x=df_test[missing_cols].isnull().sum().index, y=df_test[missing_cols].isnull().sum().values, ax=axes[1], palette='Set2')

axes[0].set_ylabel('Missing Value Count', size=15, labelpad=20)
axes[0].tick_params(axis='x', labelsize=15)
axes[0].tick_params(axis='y', labelsize=15)
axes[1].tick_params(axis='x', labelsize=15)
axes[1].tick_params(axis='y', labelsize=15)

axes[0].bar_label(axes[0].containers[0], fontsize=10)
axes[0].bar_label(axes[0].containers[1], fontsize=10)
axes[1].bar_label(axes[1].containers[0], fontsize=10)
axes[1].bar_label(axes[1].containers[1], fontsize=10)


axes[0].set_title('Training Set', fontsize=13)
axes[1].set_title('Test Set', fontsize=13)

plt.show()
```

```python
df_train["target"].value_counts()
ax = sns.barplot(x=df_train["target"].unique(), y=list(df_train["target"].value_counts()), palette='Set2')
plt.title("Target Value Counts")
ax.bar_label(ax.containers[0], fontsize=10)
ax.bar_label(ax.containers[1], fontsize=10)
axes[0].bar_label(axes[0].containers[0], fontsize=10)

plt.show()

## slightly imbalanced?
```
